```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def showImage(img):
    plt.imshow(img, cmap="gray")
```

```python
img = cv2.imread('../img/lenna.png', 0)
```

```python
showImage(img)
```

```python
#   Fourier transform
f = np.fft.fft2(img)
```

```python
f.shape
```

```python
type(f)
```

```python
type(f[0,0])
```

```python
f[0,0]
```

```python
#  Move the DC component of the FFT output to the center of the spectrum
fshift = np.fft.fftshift(f)
```

```python
fshift[0,0]
```

```python
# logarithmic transformation: 
#s = c log(1 + r); sendo 'c' uma constante e 'r' o valor do pixel antes do processamento
magnitude_spectrum = 20*np.log(np.abs(fshift))
```

```python
showImage(magnitude_spectrum)
```

```python
"""<begin>lets do inverse fourier without use filter to back to original image without loss"""
# After the frequency domain processing, then use the function np.fft.ifftshift() to perform the  
#inverse translation operation, so now the DC component returns to the upper left corner
```

```python
#and the inverse FFT inverse transform is performed using the function np.ifft2()
f_ishift = np.fft.ifftshift(fshift)
```

```python
# inverse Fourier transform
img_back_original = np.fft.ifft2(f_ishift)
```

```python
# 
img_back = np.abs(img_back)
```

```python
showImage(img_back) #aeeeee
```
