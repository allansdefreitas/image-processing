<h2>Morphological Operations</h2>

<h3>Erosion</h3>

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def showImg(img):
    plt.imshow(img, cmap="gray")
```

```python
img = cv.imread('j.png',0)
```

```python
showImg(img)
```

```python
kernel = np.ones((5,5),np.uint8) #kernel is also called 'structuring element'
kernel
```

```python
showImg(kernel)
```

```python
erosion = cv.erode(img,kernel,iterations = 1)
```

```python
showImg(erosion)
```

<h3>Dilation</h3>

```python
dilation = cv.dilate(img,kernel,iterations = 1)
```

```python
showImg(dilation)
```

<h3>Opening</h3>
<p>Erosion followed by dilation. It is useful in removing noise.</p>

```python
img = cv.imread('j_noisy.png',0)
```

```python
showImg(img)
```

```python
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
```

```python
showImg(opening)
```

<h3>Closing</h3>
<p>Dilation followed by Erosion.It is useful in closing small holes inside the foreground objects, or small black points on the object.</p> 

```python
img = cv.imread('j_hole.png',0)
```

```python
showImg(img)
```

```python
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
```

```python
showImg(closing)
```
