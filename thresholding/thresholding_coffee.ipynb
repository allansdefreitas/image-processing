```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def showImg(img):
    plt.imshow(img, cmap="gray")
```

```python
imgoriginal = cv2.imread("graos_cafe.png", 0)
```

```python
showImg(imgoriginal)
```

```python
metodo = cv2.THRESH_BINARY_INV
ret, imgbinarizada = cv2.threshold(imgoriginal, 135, 255, metodo)
```

```python
showImg(imgbinarizada)
```
